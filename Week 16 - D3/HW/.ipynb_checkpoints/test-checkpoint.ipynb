```python
import pandas as pd

from flask import Flask, render_template, jsonify, redirect, current_app

from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
from flask_sqlalchemy import SQLAlchemy
```

```python
engine = create_engine("sqlite:///db/belly_button_biodiversity.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
OTU = Base.classes.otu
SAMPLES = Base.classes.samples
META = Base.classes.samples_metadata
```

```python
session = Session(engine)
```

```python
app = Flask(__name__)
```

```python
db = SQLAlchemy(app)
```

```python
sample = 'BB_940'
```

```python
query = 'SAMPLES.' + sample
```

```python
results = session.execute("SELECT otu_id," + sample + " FROM samples ORDER BY " + sample + " DESC")
```

```python
otu_ids = []
values = []
```

```python
for result in results:
    otu_ids.append(result[0])
    values.append(result[1])
```

```python
d = {
    "otu_id" : otu_ids,
    "value": values
}
```
