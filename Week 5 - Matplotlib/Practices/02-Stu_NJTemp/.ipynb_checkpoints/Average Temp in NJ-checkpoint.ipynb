```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
months = np.arange(1,13,1)
avg_f_temp = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
plt.title("Average Temp in NJ")
plt.xlabel("Month")
plt.ylabel("Average Temp (F°)")
plt.plot(months, avg_f_temp)
plt.show()
```

```python
# calc celsius
avg_c_temp = []
for temp in avg_f_temp:
    avg_c_temp.append((temp - 32) * .56)
avg_c_temp
```

```python
plt.title("Average Temp in NJ")
plt.xlabel("Month")
plt.ylabel("Average Temp (C°)")

plt.plot(months, avg_c_temp)
plt.show()
```

```python
plt.title("Average Temp in NJ")
plt.xlabel("Month")
plt.ylabel("Average Temp (F° & C°)")

plt.plot(months, avg_f_temp)
plt.plot(months, avg_c_temp)
plt.show()
```
