# Instructions:

Read the Pokemon CSV file with Pandas.

Create a new table by extracting the following columns: "Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", and "Speed".

Find the average stats for each type of Pokemon.

Create a new DataFrame out of the averages.

calculate the total power level of each type of pokemon by adding all of the previous stats together and place the results into a new column.

Bonus:

Sort the table by strongest type and export the resulting table to a new CSV

```python
# Dependencies
import pandas as pd
import os
```

```python
# File to load
file = os.path.join('..', 'Resources',  'Pokemon.csv')
# file to write to 
write = os.path.join('..', 'output', 'pokemon_output.xlsx')
```

```python
# Read with Pandas
poke_df = pd.read_csv(file)
poke_df.head()
```

```python
# create new table with specific columns
cleaned_poke = poke_df[["Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", "Speed"]]
cleaned_poke.head()
```

```python
# group by type
grouped_poke = cleaned_poke.groupby(["Type 1"])
grouped_poke.mean().head()
```

```python
# find average stats for pokemon types
grouped_poke.mean()
# create new df for averages
average_df = pd.DataFrame(grouped_poke.mean())
average_df.mean()
```

```python
# calc total power level of each type of pokemon
average_df["Total"] = average_df["HP"] + average_df["Attack"] + average_df["Sp. Atk"] + average_df["Sp. Def"] + average_df["Speed"]
average_df["Total"]
```

```python
# sort by strongest type and export to csv
strongest_pokemon = average_df.sort_values(["Total"], ascending = False)
strongest_pokemon.reset_index(inplace = True)
strongest_pokemon.head()
```

```python
# save to excel
average_df.to_excel(write, index = True)
```
